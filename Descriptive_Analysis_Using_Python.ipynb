##                                        Descriptive Statistic Analysis
### Types of descriptive statistics
#### 1-  Measure of central tendency:   mean, median, mode, geometric mean
#### 2- Measures of variability:    range, inter quartile range, standard deviation
#### 3- Frequency distribution :   quartiles, quantiles

### Import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import House price data set

```python
df = pd.read_csv(r"D:\\datasets\\storesales\\data.csv")
```

```python
df.head()
```

### Data exploration

```python
# check the shape of data
df.shape
# 4600 records and 18 fields/columns
```

```python
df.dtypes
# there are three data types of columns i-int, ii-float, iii-object
```

```python
# find information of existing data
df.info()
```

```python
# find any null values 
df.isnull().sum()
# null/ empty values are not present
```

```python
# description of numerical columns data
df.describe()
```

```python
# find mean value of  price
df['price'].mean()
```

```python
# find mode / most occuring values of price
df['price'].mode()
```

```python
# find standard deviation of price
df['price'].std()
# dispersion / variance of sales price among different houses is :
```

```python
df['price'].min()
```

```python
df['price'].max()
# maximum price of house is :
```

```python
# find range of price
range_ = df['price'].max() - df['price'].min()
range_
# difference between min and max values of price
```

```python
# find variation of price w.r.t floors
sns.barplot(x=df['floors'],y=df['price'])
```

```python
# despersion of data 
plt.figure(figsize=(8,5))
sns.histplot(df['price'])
plt.show()
```

```python
sns.scatterplot(df['sqft_living'],df['price'])
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(df['sqft_living'],df['price'])
```

```python
# comparison between bedrooms and sqft_living
sns.barplot(df['bedrooms'],df['sqft_living'])
```
