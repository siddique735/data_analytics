```python
transactions =[]
for i in range(data.shape[0]):
    transactions.append([str(data.values[i,j]) for j in range(data.shape[1])])
transaction = np.array(transactions)    
```

```python
transaction
```

```python
te = TransactionEncoder()
tr_ary=te.fit_transform(transaction)
```

```python
tr_ary
```

```python
dataset=pd.DataFrame(tr_ary, columns=te.columns_)
```

```python
dataset.columns
```

```python
def data_encode(x):
    if x == False:
        return 0
    else:
        return 1
    
dataset=dataset.applymap(data_encode)    
```

```python
dataset
```

#### implement algorithm

```python
model=apriori(dataset,min_support=0.05,use_colnames=True)
```

```python
model['length']=model['itemsets'].apply(lambda x: len(x))
```

```python
model
```

```python
#explore items with length 2
```

```python
model[(model['length']==2)& (model['support'] >=0.05)]
```

```python
model[(model['length']==2)& (model['support'] >=0.8)]
```

```python
# extract itemsets with rule creation
```

```python
rules = association_rules(model,metric='lift',min_threshold=1.1)
```

```python
rules
```

```python
rules["antecedents_length"]=rules["antecedents"].apply(lambda x:len(x))
rules["consequents_length"]=rules["consequents"].apply(lambda x:len(x))
rules.sort_values('lift',ascending=False)
```

```python
rules=rules.sort_values('confidence',ascending=False)
```
