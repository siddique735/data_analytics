```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.ambitionbox.com/list-of-companies'
```

```python
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.content,'html5lib')
```

```python
#soup.prettify()
```

```python
company=soup.find_all('div', class_='company-info-wrapper')
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p').text.strip())
    reviews.append(i.find('a', class_='review-count').text.strip())
    ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    
    
    
    
```

```python
df=pd.DataFrame({'name':name,'rating':rating,'reviews':reviews,'ctype':ctype})
```
